```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#from scipy.stats import f_oneway
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
import numpy as np
#!pip install researchpy
#import researchpy as rp
# @ load data

file_path='/content/drive/MyDrive/University/Quinto_Semestre/LAB-Arqui-compu/MatascaBrian-GaviriaEsteban.xlsx'
# df = pd.read_excel(file_path,'Hoja 1')
df = pd.read_excel(file_path,'At√≠picos Eliminados')

df = df.rename(columns={"Normalized (ns)":"Normalized_ns"})
df['version'] = pd.Categorical(df['version'])
df.info()

# @ summary data
print("//////////////// summary data  ////////////////////////////////")
print(df.head())
Z=1.96
#E=np.sqrt()
Stat_data=df[["Normalized_ns", "version"]].groupby("version").agg({"Normalized_ns":["min", "max", "median", "mean" ,np.std,np.var]})
Stat_data=Stat_data.Normalized_ns
print(Stat_data)
Error=Stat_data['mean']/100  # 1%
print("")
print("error (ns) al 1%")
print(Error)

print("Tama√±o de muestra")
print(Stat_data['var']*(Z*Z)/(Error*Error))
#print(rp.summary_cont(df['Normalized_ns']))
#print(rp.summary_cont(df['Normalized_ns'].groupby(df['version'])))
print(" **")

# "//////////////// Gr√°fico de Cajas y Bigotes ////////////////////////////////")
sns.set(style="whitegrid")  # Establecer el estilo del gr√°fico
plt.figure(figsize=(10, 6))  # Tama√±o del gr√°fico

# Crear el gr√°fico de cajas y bigotes
sns.boxplot(x='version', y='Normalized_ns', data=df)
# Etiquetas y t√≠tulo
plt.xlabel('Algoritmo')
plt.ylabel('Tiempo normalizado (ns)')
plt.title('Tiempos de Ejecuci√≥n')
# Mostrar el gr√°fico
plt.show()
print("")

modelo = ols('Normalized_ns ~ C(version)', alpha=0.05, data=df).fit()
# Mostrar los resultados
print("===========          TABLA ANOVA         ================")
tabla_anova = anova_lm(modelo, typ=2) # Type 2 ANOVA DataFrame
print(tabla_anova)
print(" ")
print("Estad√≠stica de prueba F:", modelo.fvalue)
print("Valor p:", modelo.f_pvalue)
print("alpha: 0.05")
print("================= -------------- =========================")
print("")

print("======================================     filtrar datos Ver4 ver6  ==============================")
df_v4_6 = df[df['version'].isin(['ver4', 'ver6'])]
print(df_v4_6.head())
df_v4_6.info()

# Crear el gr√°fico de cajas y bigotes
sns.boxplot(x='version', y='Normalized_ns', data=df_v4_6)
# Etiquetas y t√≠tulo
plt.xlabel('Algoritmo')
plt.ylabel('Tiempo normalizado (ns)')
plt.title('Tiempos de Ejecuci√≥n v4 v6 ')
# Mostrar el gr√°fico
plt.show()
print("")
#print(rp.summary_cont(df_v2_5['Normalized_ns']))
print("")
#print(rp.summary_cont(df_v2_5['Normalized_ns'].groupby(df_v4_6['version'])))
print("===========          ANOVA Ver4 - ver6     ================")
modelo4_6 = ols('Normalized_ns ~ C(version)', alpha=0.05, data=df_v4_6).fit()
tabla_anova = anova_lm(modelo4_6, typ=2) # Type 2 ANOVA DataFrame
print(tabla_anova)
print(" ")
print("Estad√≠stica de prueba F:", modelo4_6.fvalue)
print("Valor p:", modelo4_6.f_pvalue)
print("alpha: 0.05")

print("================= ----------------- =========================")
print("")
print("==========================================     filtrar datos Ver2 ver5  ==================================")
df_v2_5 = df[df['version'].isin(['ver2', 'ver5'])]
print(df_v2_5.head())
df_v2_5.info()

# "//////////////// Gr√°fico de Cajas y Bigotes ////////////////////////////////")
sns.set(style="whitegrid")  # Establecer el estilo del gr√°fico
plt.figure(figsize=(10, 6))  # Tama√±o del gr√°fico

# Crear el gr√°fico de cajas y bigotes
sns.boxplot(x='version', y='Normalized_ns', data=df_v2_5)
# Etiquetas y t√≠tulo
plt.xlabel('Algoritmo')
plt.ylabel('Tiempo normalizado (ns)')
plt.title('Tiempos de Ejecuci√≥n v2 v5')
# Mostrar el gr√°fico
plt.show()
print("")
print("===========          ANOVA Ver2 - ver5   ================")
modelo2_5 = ols('Normalized_ns ~ C(version)', alpha=0.05, data=df_v2_5).fit()
tabla_anova = anova_lm(modelo2_5, typ=2) # Type 2 ANOVA DataFrame
print(tabla_anova)
print(" ")
print("Estad√≠stica de prueba F:", modelo2_5.fvalue)
print("Valor p:", modelo2_5.f_pvalue)
print("alpha: 0.05")
print("================= ----------------- =========================")
print("")

print("")
print("==========================================     filtrar datos Ver1 ver3  ==================================")
df_v1_3 = df[df['version'].isin(['ver1', 'ver3'])]
print(df_v1_3.head())
df_v1_3.info()

# "//////////////// Gr√°fico de Cajas y Bigotes ////////////////////////////////")
sns.set(style="whitegrid")  # Establecer el estilo del gr√°fico
plt.figure(figsize=(10, 6))  # Tama√±o del gr√°fico

# Crear el gr√°fico de cajas y bigotes
sns.boxplot(x='version', y='Normalized_ns', data=df_v1_3)
# Etiquetas y t√≠tulo
plt.xlabel('Algoritmo')
plt.ylabel('Tiempo normalizado (ns)')
plt.title('Tiempos de Ejecuci√≥n v1 v3')
# Mostrar el gr√°fico
plt.show()
print("")
print("===========          ANOVA Ver1 - ver3   ================")
modelo1_3 = ols('Normalized_ns ~ C(version)', alpha=0.05, data=df_v1_3).fit()
tabla_anova = anova_lm(modelo1_3, typ=2) # Type 2 ANOVA DataFrame
print(tabla_anova)
print(" ")
print("Estad√≠stica de prueba F:", modelo1_3.fvalue)
print("Valor p:", modelo1_3.f_pvalue)
print("alpha: 0.05")
print("================= ----------------- =========================")
print("")
#print("===================================     filtrar datos Ver1 ver3  samples 5 - 29 ============================")
#nuevo_df_v2_5 = df[(df['version'].isin(['ver2', 'ver5'])) & (df['Sample'].between(5, 29))]
#print(nuevo_df_v2_5.head())
#nuevo_df_v2_5.info()
## @ summary data
#print("//////////////// summary data  ////////////////////////////////")
#print(nuevo_df_v2_5[["Normalized_ns", "version"]].groupby("version").agg({"Normalized_ns":["min", "max", "median", "mean",np.var]}))


## "//////////////// Gr√°fico de Cajas y Bigotes ////////////////////////////////")
#sns.set(style="whitegrid")  # Establecer el estilo del gr√°fico
#plt.figure(figsize=(10, 6))  # Tama√±o del gr√°fico

## Crear el gr√°fico de cajas y bigotes
#sns.boxplot(x='version', y='Normalized_ns', data=nuevo_df_v2_5)
## Etiquetas y t√≠tulo
#plt.xlabel('Algoritmo')
#plt.ylabel('Tiempo normalizado (ns)')
#plt.title('Gr√°fico de Cajas y Bigotes para Tiempos de Ejecuci√≥n v4 v6 sample 5-29 ')
# Mostrar el gr√°fico
#plt.show()
#print("")

print("")
print("END")


```
